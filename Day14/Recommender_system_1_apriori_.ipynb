```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import os
os.chdir(r"C:\Users\dai\Desktop\S-AI\MachineLearning\Day12\Datasets")
```

```python
fp_df = pd.read_csv('Faceplate.csv',index_col=0)
fp_df = fp_df.astype(bool)
fp_df
```

```python
itemsets = apriori(fp_df, min_support=0.2, use_colnames=True)
itemsets
```

```python
# itemsets = apriori(fp_df, min_support=0.2, use_colnames=False)
# itemsets
```

```python
# itemsets = apriori(fp_df, min_support=0.5, use_colnames=True)
# itemsets
```

```python
rules = association_rules(df =itemsets, num_itemsets=len(fp_df), metric='lift',
                          min_threshold=1,return_metrics=['support','confidence','lift'])
rules.sort_values('confidence',ascending = False)
```
