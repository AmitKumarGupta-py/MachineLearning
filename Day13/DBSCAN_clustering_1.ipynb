## -1 are not added in cluster they are outliers

```python
eps_range = [0.2,0.4,0.6,1]
mp_range = [2,3,4,5]
cnt = 0
a =[]
for i in eps_range:
    for j in mp_range:
        clust_DB = DBSCAN(eps=i, min_samples=j)
        clust_DB.fit(milk_scaled.iloc[:,:5])
        if len(set(clust_DB.labels_)) > 2: ## length of the set should be greater than 2 for silhoutte score 
            cnt = cnt + 1
            milk_scaled['Clust'] = clust_DB.labels_
            milk_scl_inliers = milk_scaled[milk_scaled['Clust']!=-1]## filtering the outliers
            sil_sc = silhouette_score(milk_scl_inliers.iloc[:,:-1],# dataframe # last column not consider
                             milk_scl_inliers.iloc[:,-1])# label # the last column
            a.append([cnt,i,j,sil_sc])
            print(i,j,sil_sc)
    
a = np.array(a)
pa = pd.DataFrame(a,columns=['Sr','eps','min_pt','sil'])
print("Best Paramters:")
pa[pa['sil'] == pa['sil'].max()]
```
