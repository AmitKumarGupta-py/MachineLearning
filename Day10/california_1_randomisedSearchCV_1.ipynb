```python
from sklearn.datasets import fetch_california_housing

X,y = fetch_california_housing(as_frame=True, return_X_y = True)

X.columns
```

```python
X.info(),y.info()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=24,test_size=0.3)
```

```python
rf = RandomForestRegressor( random_state=24, n_jobs = -1)#n_jobs= -1 will make execution fast using all core of cpu
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
r2_score(y_test,y_pred)
```

```python
params = {'max_features':[2,3,4,5,6,None],
         'max_depth':[None, 3,5],
         'min_samples_split':[2,10,50],
          'min_samples_leaf':[1,10,50],
          'n_estimators':[20,50]
         }

kfold = KFold(n_splits = 5,shuffle = True,random_state =24)
```

```python
gcv = GridSearchCV(rf,param_grid = params, cv = kfold, scoring = 'r2',verbose = 3)
gcv.fit(X,y)
```

```python
best_model = gcv.best_estimator_
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
