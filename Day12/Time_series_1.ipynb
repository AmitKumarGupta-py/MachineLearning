```python
import pandas as pd
import numpy as np
import os 
os.chdir(r"C:\Users\dai\Desktop\S-AI\MachineLearning\Day12\Datasets")
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
passengers = pd.read_csv("AirPassengers.csv",index_col=0)
```

```python
passengers.plot()
plt.title("Monthly passengers")
plt.show()
```

```python
passengers.index = pd.DatetimeIndex(passengers.index).to_period("M")
```

```python
type(passengers.index)
```

```python
q_pass = passengers.resample('Q').sum()
q_pass.index.rename('Quarter', inplace=True)
q_pass.plot()
plt.title('Quaterly Passengers')
plt.show()
```

```python
a_pass = passengers.resample('Y').sum() ## ''Y'' can take 
a_pass.index.rename('Annual', inplace=True)
a_pass.plot()
plt.title('Annual Passengers')
plt.show()
```

```python
passengers = pd.read_csv("AirPassengers.csv",index_col=0)
result = seasonal_decompose(passengers['Passengers'],model='additive',period=12)

result.plot()
# fig_size = plt.rcParams["figure.figsize"]
# fig_size[0] = 6
# fig_size[1] = 4
  
plt.title('Additive Decomposition')

plt.show()
```

```python

result = seasonal_decompose(passengers['Passengers'],model='multiplicative',period=12)
result.plot()
plt.title('Multiplicative Decomposition')
plt.show()
```

```python
y = passengers['Passengers']
```

## Centered Moving average (rOLLING Centered Mean)

```python
span = 3
fcast = y.rolling(span, center=True).mean()
plt.plot(y, label='Data')
plt.plot(fcast,label='Cantered Rolling mean')
plt.legend(loc='best')
plt.show()
```
